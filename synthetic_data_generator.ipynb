```python
import numpy as np
import pandas as pd
from scipy.special import expit  # Logistic function
```

```python
# Sample number
n_samples = 5000
np.random.seed(42)

# Simulated variables
humidity = np.clip(np.random.normal(70, 10, n_samples), 40, 95)        # %
temperature = np.random.uniform(10, 25, n_samples)                     # °C
sun_hours = np.random.uniform(3, 10, n_samples)                        # horas
crop_age = np.random.uniform(30, 120, n_samples)                       # días
isolation_index = np.random.beta(2, 3, n_samples)                      # 0–1
precipitation = np.clip(np.random.gamma(2, 3, n_samples), 0, 40)       # mm
humidity_var = np.clip(np.random.normal(0, 8, n_samples), -20, 20)     # % últimos 3 días
temp_var = np.clip(np.random.normal(0, 5, n_samples), -15, 15)         # °C últimos 3 días
```

```python
# Infection Probability Logistic Function
logit = (
    -8.0
    + 0.08 * humidity
    + 0.06 * precipitation
    + 0.04 * crop_age
    - 2.5 * isolation_index
    - 0.10 * sun_hours
    + 0.03 * humidity_var
    + 0.02 * temp_var
)

infection_prob = expit(logit)
label = (infection_prob > 0.5).astype(int)
```

```python
df = pd.DataFrame({
    'humidity': humidity,
    'temperature': temperature,
    'sun_hours': sun_hours,
    'crop_age': crop_age,
    'isolation_index': isolation_index,
    'precipitation': precipitation,
    'humidity_var': humidity_var,
    'temp_var': temp_var,
    'infection_prob': infection_prob,
    'label': label
})

df.head()
```

```python
df.to_excel('/content/potato_disease_dataset.xlsx', index=False)
df.to_csv('/content/potato_disease_dataset.csv', index=False)
print("Dataset guardado en Colab: potato_disease_dataset.xlsx y potato_disease_dataset.csv")
```
